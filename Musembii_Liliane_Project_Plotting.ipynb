```python
# YOUR CODE HERE
plt.figure(figsize=(10, 6))
plt.hist(df['co2'], bins=20, color='blue', edgecolor='black')
plt.title('Histogram of Carbon Dioxide Emissions')
plt.xlabel('CO2 Emissions (g/mile)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# YOUR CODE HERE
plt.figure(figsize=(10, 6))
plt.hist(df['co2'], bins=10, color='blue', edgecolor='black')
plt.title('Histogram of Carbon Dioxide Emissions')
plt.xlabel('CO2 Emissions (g/mile)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# YOUR CODE HERE
plt.figure(figsize=(10, 6))
plt.hist(df['comb'], bins=20, color='green', edgecolor='black')  # Histogram for combined fuel economy
plt.title('Histogram of Combined Fuel Economy')
plt.xlabel('Fuel Economy (MPG)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.hist(df['feScore'], bins=20, color='red', edgecolor='black')  # Histogram for fuel economy score
plt.title('Histogram of Fuel Economy Score')
plt.xlabel('Fuel Economy Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


# Load the dataset into the 'df' variable
df = pd.read_csv('https://drive.google.com/uc?id=16lgiSKILxHaLqmNUkCafQIAvyL2Y_KYf')

plt.figure(figsize=(14, 10))
correlation = df.corr(numeric_only=True)  # Only use number columns
sns.heatmap(correlation, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

```

```python
#Interpretation
# There's a strong positive correlation close to +1 that shows two variables increase together
# Fuel consumption and CO2 Emissions are are strongly positively correlated.
# There's a strong negative correlation, close to -1, that indicates that one increases as the other decreases. Smog Rating may negatively correlate with CO2 Emissions
# Ghg Score has a negative correlation with CO2 Emissions and Fuel Consumption, meaning higher GHG scores are associated with more environmentally friendly.

#Finally What Columns would you delete Drop if the target is the ghg score? Justify your Answer
#vehicle_class	Can be informative, but may be redundant with weight or engine size
#smog_rating	Strongly correlated with ghg_score — could lead to data leakage if included
```
